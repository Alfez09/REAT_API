```python
import requests
import os

def get_place_photos(fsq_id, limit=10, sort='POPULAR'):
    url = f"https://api.foursquare.com/v3/places/{fsq_id}/photos"
    api_key = "fsq3FGQsZS8l/x9Pg1wWRDNPbqwL2JUpVhnSZVt5eOyPh1g="
    
    headers = {
        "accept": "application/json",
        "Authorization": api_key
    }
    param = {
        'limit':limit,
        'sort':sort
    }
    
    resp = requests.get(url, params=param, headers=headers)
    if resp.status_code == 200:
        data = resp.json()
        current_dir = os.getcwd()
        folder_name = input('Enter folder name: ')
        os.mkdir(folder_name)
        new_path = os.path.join(current_dir, folder_name)
        os.chdir(new_path)
        for i in range(len(data)):
            prefix = data[i]['prefix']
            suffix = data[i]['suffix']
            img_url = prefix + 'original' + suffix
            img_resp = requests.get(img_url)
            if img_resp.status_code == 200:
                with open(f'img_{i+1}.jpeg', 'wb') as f:
                    f.write(img_resp.content)
        else:
            os.chdir(current_dir)
            print(f"Images are downloaded...")
    else:
        print(f"Error occured status code {resp.status_code} Reason: {resp.reason}")
```
